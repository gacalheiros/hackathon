### How to define failure

Let's check how the number of disks change by datetime for different experiments

```python
plot_data_cms   = data[data.exp=='cms'][['Serial','date']].groupby('date').Serial.nunique()
plot_data_atlas = data[data.exp=='atlas'][['Serial','date']].groupby('date').Serial.nunique()
plot_data_lhcb  = data[data.exp=='lhcb'][['Serial','date']].groupby('date').Serial.nunique()
plot_data_alice = data[data.exp=='alice'][['Serial','date']].groupby('date').Serial.nunique()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(plot_data_cms.index, plot_data_cms,     s=3, c='r', label='cms')
plt.scatter(plot_data_atlas.index, plot_data_atlas, s=3, c='g', label='atlas')
plt.scatter(plot_data_lhcb.index, plot_data_lhcb,   s=3, c='b', label='lhcb')
plt.scatter(plot_data_alice.index, plot_data_alice, s=3, c='m', label='alice')
plt.xlabel("date")
plt.ylabel("number of disks")
plt.title('Timeline of number of disks used per experiment')
plt.legend(loc='best')
plt.xlim("2017-10-01 02:00:00","2017-10-21 03:00:00")
plt.show()
```

If a disk disapears from the timeline, we consider it as a failure moment;
The data from the failure moment seven days ago will be labeled as 0 (which means possible failure).

From the diagram above, we find that from "2017-10-01 02:00:00" to "2017-10-21 03:00:00" all disks disappear. Thus, for any disk disappear in this period of time, we do not consider it as a failure.

### Find the failure moments

#### First, find the possible bad hosts

```python
# let's identify amount of serial per day for each host
host_date_number = data.groupby(['host', 'date'])['Serial'].count()
```

```python
# calculate std for timeseries mentioned above
get_std_for_ts = host_date_number.groupby('host').std()

# let's grab all host with non-constant std, which means that they have some changes in amount of disks during period of usage
# these are the host id's where a disk failure might have occured
possibly_bad_hosts = get_std_for_ts[get_std_for_ts!=0]
print(f'Amount of hosts with possibly failured hosts: {len(possibly_bad_hosts)}')
```

```python
# let's take subset for only possibly failed hosts
possibly_bad_hosts_df = data[['host','Serial','date']][data.host.isin(possibly_bad_hosts.index)].groupby(['host','date'])['Serial'].apply(list).reset_index() 
# now we get a dataframe where the number of unique serials is collected in a list for every host & date
possibly_bad_hosts_df.head()
```

#### Second, find possible bad disks

```python
# let's find our bad drives (you can and should improve this method)
bad_serials = []
bad_moments = {}
date = "2017-10-21 03:00:00"

# loop over every host with possibility of a hard disk failure 
for host in possibly_bad_hosts.index:
    
    current_host_df = possibly_bad_hosts_df[possibly_bad_hosts_df.host == host]
    prev_set = set([])
    # if the number of hard disk reporting for a given date is one less than the day before we label it as a failure
    row_iterator = current_host_df.iterrows()
    _, last = next(row_iterator)
    for idx, row in row_iterator:
        current_set = set(row[2])
        difference  = prev_set.difference(current_set)
        next_set = set(last[2])
        #Comparison to the next row to see whether the same serial number get put back. If it is the same then it does not count.
        last = row
        for item in next_set:
            if item in bad_serials:
                bad_serials.remove(item) 
        if  prev_set.issubset(current_set) or len(current_set) >= len(prev_set):
            prev_set = current_set
            continue
        elif len(difference) == 1 and row.date!=date:
            # get failed serial 
            key = difference.pop()
            bad_serials.append(key)
            bad_moments[key] = row.date
            prev_set = current_set
        elif len(difference) == 2 and row.date!=date:
            # get failed serial 
            key = difference.pop()
            key2 = difference.pop()
            bad_serials.append(key)
            bad_serials.append(key2)
            bad_moments[key] = row.date
            bad_moments[key2] = row.date
            prev_set = current_set
        elif len(difference) == 3 and row.date!=date:
            # get failed serial 
            key = difference.pop()
            key2 = difference.pop()
            key3 = difference.pop()
            bad_serials.append(key)
            bad_serials.append(key2)
            bad_serials.append(key3)
            bad_moments[key] = row.date
            bad_moments[key2] = row.date
            bad_moments[key3] = row.date
            prev_set = current_set
        elif len(difference) == 4 and row.date!=date:
            # get failed serial 
            key = difference.pop()
            key2 = difference.pop()
            key3 = difference.pop()
            key4 = difference.pop()
            bad_serials.append(key)
            bad_serials.append(key2)
            bad_serials.append(key3)
            bad_serials.append(key4)
            bad_moments[key] = row.date
            bad_moments[key2] = row.date
            bad_moments[key3] = row.date
            bad_moments[key4] = row.date
            prev_set = current_set
```

```python
print(f'Amount of failured disks: {len(bad_serials)}')
```

In the test above, we find that there is no difference larger than 4 and the number of failure diaks is maximumly 52.

### Label the failure period (from failure moment to 7 days ago) of each failure disk

#### Data normalization

```python
feature_names = ["smart1", 'smart2', 'smart3', 'smart4', 'smart5', 'smart7', 'smart8', 'smart9', 'smart10', 'smart12', 'smart192', 'smart193', 'smart194', 'smart196', 'smart197', 'smart198', 'smart199']
```

```python
# Data normalization
def normalize_input(x):
    normalized = (x-min(x))/(max(x)-min(x))
    
    return normalized
```

```python
for feature in feature_names:
    data[feature] = normalize_input(data[feature].values)
```

```python
failed_disks = data[data.Serial.isin(bad_serials)]
```

```python
failed_disks
```

```python
print(f'First ten bad disks:\n{failed_disks.Serial.unique()[:10]}')
```

For each bad serial, we only take the data within the failure period

```python
X_bad = []

for cur_disk, value in bad_moments.items():
    cur_moment = value
    cur_fail_disk = failed_disks[failed_disks.Serial == cur_disk]
    pre_moment = pd.date_range(end = cur_moment, periods=8, freq='D')[0]

    tmp_1d_list = []
    tmp_2d_list = []

    if cur_fail_disk.date.min() < pre_moment:
        cur_fail_period = cur_fail_disk[(cur_fail_disk['date'] <= cur_moment) & (cur_fail_disk['date'] >= (pre_moment))]
        for fail_feature in feature_names:
            tmp_1d_list = cur_fail_period[fail_feature].values.ravel()
            tmp_2d_list.append(np.array(tmp_1d_list))
        tmp_2d_list = np.array(tmp_2d_list)
        if(tmp_2d_list.shape==(17, 7)):
            X_bad.append(np.array(tmp_2d_list))
    else:
        pass
```

```python
print("We have", len(X_bad), "failure data lists,", len(X_bad[0]), "features,", len(X_bad[0][0]), "days.")
```

```python
X_bad = np.dstack(X_bad)
```

```python
X_bad.shape
```

```python
X_bad=np.rollaxis(X_bad,-1)
print(X_bad.shape)
```

```python
X_bad = X_bad.astype(np.float32)
```

```python
Y_bad=np.ones(34).astype(np.float32)
```
