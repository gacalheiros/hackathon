```python
import pandas  as pd
import numpy   as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = pd.read_csv('smart_data_clean.csv')
print('data loaded')
```

```python
data_cms = data[data.exp=='cms']
data_cms.head(10)
```

```python
data_cms.drop(columns=['exp'])
```

```python
data_cms.shape
```

```python
plot_data_cms   = data_cms[['Serial','date']].groupby('date').Serial.nunique()
plt.figure(figsize=(15,7))
plt.scatter(plot_data_cms.index, plot_data_cms,     s=3, c='r', label='cms')

plt.xlabel("date")
plt.ylabel("number of disks")
plt.title('Timeline of number of disks used per experiment')
plt.legend(loc='best')
plt.show()
```
